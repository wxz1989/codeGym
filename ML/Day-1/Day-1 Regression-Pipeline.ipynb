```python
import pandas as pd
```

```python
df = pd.read_csv("insurance.csv")
```

```python
from sklearn import model_selection
from sklearn import linear_model
from sklearn import pipeline, preprocessing
```

```python
#choosing target column as Charges Column
target  =  "charges" 

# from Source Columns remove the Charges Column
X = df.drop(columns=[target])
X = pd.get_dummies(X, drop_first=True)
columns = X.columns
y = df[target]
X_train, X_test, y_train, y_test = model_selection.train_test_split(X.values, y.values
                                    , test_size = 0.3, random_state = 1)

pipe = pipeline.Pipeline([
    ("scaler", preprocessing.StandardScaler()),
    ("est", linear_model.LinearRegression())
])

pipe.fit(X_train, y_train)
```

```python
est = pipe.steps[-1][-1]
```

```python
est.coef_
```

```python
y_train_pred = pipe.predict(X_train)
```

```python
y_test_pred = pipe.predict(X_test)
```

```python
result = pd.DataFrame({"true": y_test, "predicted": y_test_pred})
result["error"] = result["true"] - result["predicted"]
```

```python
result
```

```python
import numpy as np
```

```python
#Computing Mean Squared Error
mse = np.mean(result.error ** 2) # Lower is better
mse
```

```python
y_test_var = np.var(y_test)
mse/y_test_var
mse / np.mean((y_test - y_train.mean()) ** 2)
r2 = 1 - mse / y_test_var
r2
```
